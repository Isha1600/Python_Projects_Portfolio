## Create storage.py

### Subtask:
Create a `storage.py` file containing `save_data` and `load_data` functions using `pickle` for serializing and deserializing student data.


**Reasoning**:
To fulfill the subtask, I will create the `storage.py` file with the `save_data` and `load_data` functions as specified in the instructions, utilizing the `pickle` module for data persistence.



```python
%%writefile storage.py

import pickle

def save_data(data, filename):
    """Saves data to a file using pickle."""
    with open(filename, 'wb') as f:
        pickle.dump(data, f)
    print(f"Data successfully saved to {filename}")

def load_data(filename):
    """Loads data from a file using pickle."""
    try:
        with open(filename, 'rb') as f:
            data = pickle.load(f)
        print(f"Data successfully loaded from {filename}")
        return data
    except FileNotFoundError:
        print(f"File '{filename}' not found. Returning an empty list.")
        return []
```

## Integrate Persistence into manager.py

### Subtask:
Rewrite `manager.py` to incorporate the `storage.py` functions by adding `save_students` and `load_students` methods to the `StudentManager` class. These methods will handle saving and loading the list of students.


**Reasoning**:
To integrate persistence into `manager.py`, I need to rewrite the file using `%%writefile` and add the `save_students` and `load_students` methods to the `StudentManager` class. These methods will import and utilize the `save_data` and `load_data` functions from `storage.py` to handle the serialization and deserialization of student data.



```python
%%writefile manager.py

from student import Student
from storage import save_data, load_data

class StudentManager:
    def __init__(self):
        self.students = []

    def add_student(self, student_id, name, grades):
        student = Student(student_id, name, grades)
        self.students.append(student)

    def display_students(self):
        if not self.students:
            print("No students in the system.")
        else:
            print("\n--- Student Records ---")
            for student in self.students:
                print(student)
            print("-----------------------")

    def save_students(self, filename):
        save_data(self.students, filename)

    def load_students(self, filename):
        self.students = load_data(filename)
```

```python
%%writefile main.py

from manager import StudentManager
import os

def main():
    filename = "students.pkl"
    manager = StudentManager()

    # Try to load existing data first
    print("--- Attempting to load existing students ---")
    manager.load_students(filename)
    print("--- Current students after load (if any) ---")
    manager.display_students()

    # Add a combination of previous data and new data to reach at least 10 students
    print("\n--- Adding new students records ---")
    # Ensure unique student IDs
    current_student_ids = {s.student_id for s in manager.students}

    if "S001" not in current_student_ids:
        manager.add_student("S001", "Alice Smith", [90, 85, 92])
    if "S002" not in current_student_ids:
        manager.add_student("S002", "Bob Johnson", [78, 88, 80])
    if "S003" not in current_student_ids:
        manager.add_student("S003", "Isha", [85, 60, 90])
    if "S004" not in current_student_ids:
        manager.add_student("S004", "Tina", [30, 55, 70])
    if "S005" not in current_student_ids:
        manager.add_student("S005", "Rahul", [60, 45, 60])
    if "S006" not in current_student_ids:
        manager.add_student("S006", "Priya", [95, 90, 98])
    if "S007" not in current_student_ids:
        manager.add_student("S007", "Amit", [70, 65, 72])
    if "S008" not in current_student_ids:
        manager.add_student("S008", "Sara", [88, 82, 90])
    if "S009" not in current_student_ids:
        manager.add_student("S009", "Tom", [55, 60, 58])
    if "S010" not in current_student_ids:
        manager.add_student("S010", "Jerry", [75, 70, 78])


    print("\n--- All students ---")
    manager.display_students()

    # Save the combined data
    print(f"\n--- Saving to {filename} ---")
    manager.save_students(filename)

    # Simulate fresh start and load again to verify persistence
    print("\n--- Creating new manager instance and loading data ---")
    new_manager = StudentManager()
    new_manager.load_students(filename)

    print("\n--- Students loaded into new manager to verify persistence ---")
    new_manager.display_students()

    # Clean up the created pickle file
    if os.path.exists(filename):
        os.remove(filename)
        print(f"\nCleaned up {filename}")

if __name__ == "__main__":
    main()
```

## Rewrite student.py

### Subtask:
Rewrite the `student.py` file to define the `Student` class with `name` and a single `marks` attribute, along with methods for `get_name`, `get_marks`, `update_marks`, `is_pass` (based on `marks`), and `__str__` to include pass/fail status. This aligns with the user's implied structure and the `add_student` calls provided.


**Reasoning**:
To fulfill the subtask, I need to rewrite the `student.py` file with the updated `Student` class definition, including the new methods and `marks` attribute, using the `%%writefile` magic command.



```python
%%writefile student.py

class Student:
    def __init__(self, student_id, name, marks):
        self.student_id = student_id
        self.name = name
        self.marks = marks

    def get_name(self):
        return self.name

    def get_marks(self):
        return self.marks

    def update_marks(self, new_marks):
        self.marks = new_marks

    def is_pass(self):
        return self.marks >= 50

    def __str__(self):
        status = "Passed" if self.is_pass() else "Failed"
        return f"Student ID: {self.student_id}, Name: {self.name}, Marks: {self.marks}, Status: {status}"
```

## Rewrite manager.py

### Subtask:
Rewrite `manager.py` to update the `StudentManager` class. This update should include the following functionalities:
1.  **Persistence**: Ensure `save_students` and `load_students` methods are present and correctly utilize `storage.py`.
2.  **`remove_student(self, student_id)`**: A method to remove a student by their ID.
3.  **`update_student(self, student_id, new_name=None, new_marks=None)`**: A method to update a student's name or marks by their ID.
4.  **`display_all(self)`**: A method to display all students, using the updated `__str__` of the `Student` class.
5.  **`show_passed(self)`**: A method to display only the students who have passed (marks >= 50).

**Reasoning**:
To fulfill the subtask, I need to rewrite the `manager.py` file using `%%writefile` to include the `StudentManager` class with the specified new methods (`remove_student`, `update_student`, `display_all`, `show_passed`) and ensure persistence methods (`save_students`, `load_students`) are correctly integrated.



```python
%%writefile manager.py

from student import Student
from storage import save_data, load_data

class StudentManager:
    def __init__(self):
        self.students = []

    def add_student(self, student_id, name, marks):
        # Check if student_id already exists
        if any(s.student_id == student_id for s in self.students):
            print(f"Error: Student with ID {student_id} already exists.")
            return
        student = Student(student_id, name, marks)
        self.students.append(student)
        print(f"Student {name} (ID: {student_id}) added.")

    def remove_student(self, student_id):
        initial_len = len(self.students)
        self.students = [s for s in self.students if s.student_id != student_id]
        if len(self.students) < initial_len:
            print(f"Student with ID {student_id} removed.")
        else:
            print(f"Student with ID {student_id} not found.")

    def update_student(self, student_id, new_name=None, new_marks=None):
        found = False
        for student in self.students:
            if student.student_id == student_id:
                if new_name is not None:
                    student.name = new_name
                if new_marks is not None:
                    student.update_marks(new_marks)
                found = True
                print(f"Student with ID {student_id} updated: Name='{student.name}', Marks='{student.marks}'.")
                break
        if not found:
            print(f"Student with ID {student_id} not found.")

    def display_all(self):
        if not self.students:
            print("No students in the system.")
        else:
            print("\n--- All Student Records ---")
            for student in self.students:
                print(student)
            print("---------------------------")

    def show_passed(self):
        passed_students = [s for s in self.students if s.is_pass()]
        if not passed_students:
            print("No students passed.")
        else:
            print("\n--- Passed Student Records ---")
            for student in passed_students:
                print(student)
            print("------------------------------")

    def save_students(self, filename):
        save_data(self.students, filename)

    def load_students(self, filename):
        loaded_data = load_data(filename)
        # Ensure we only load valid Student objects
        if isinstance(loaded_data, list) and all(isinstance(item, Student) for item in loaded_data):
            self.students = loaded_data
        else:
            print("Loaded data is not in the expected format (list of Student objects). Initializing with empty list.")
            self.students = []
```

## Rewrite main.py

### Subtask:
Rewrite `main.py` to demonstrate the full functionality of the enhanced student management system. This will include:
1.  **Initializing `StudentManager` and loading existing data.**
2.  **Adding students**: Add at least 10 student records, ensuring a mix of passing and failing marks, and preventing duplicate `student_id`s.
3.  **Displaying all students** using the new `display_all` method.
4.  **Updating a student's record** (e.g., change name or marks).
5.  **Removing a student's record**.
6.  **Displaying only passed students** using the new `show_passed` method.
7.  **Saving the updated student data**.
8.  **Loading data into a new manager instance** to verify persistence.
9.  **Displaying all students from the new manager instance**.
10. **Cleaning up** the persistence file (`students.pkl`).

**Reasoning**:
To fulfill the subtask of rewriting `main.py`, I will create a new `main.py` file using `%%writefile`. This file will import `StudentManager` and `os`, and define a `main` function that demonstrates all the requested functionalities: initializing the manager, loading data, adding students, displaying all students, updating a student, removing a student, displaying passed students, saving data, loading data into a new manager, displaying again, and finally cleaning up the persistence file. The `add_student` calls will include a mix of passing and failing marks and will handle potential duplicates.



```python
%%writefile main.py

from manager import StudentManager
import os

def main():
    filename = "students.pkl"
    manager = StudentManager()

    print("\n--- Initializing and attempting to load existing students ---")
    manager.load_students(filename)
    print("--- Students after initial load (if any) ---")
    manager.display_all()

    # Adding students
    print("\n--- Adding new students records ---")
    # Using a set to track current student IDs to avoid adding duplicates if loaded from file
    current_student_ids = {s.student_id for s in manager.students}

    students_to_add = [
        ("S001", "Alice Smith", 90),
        ("S002", "Bob Johnson", 45),
        ("S003", "Isha", 85),
        ("S004", "Tina", 30),
        ("S005", "Rahul", 60),
        ("S006", "Priya", 95),
        ("S007", "Amit", 70),
        ("S008", "Sara", 48),
        ("S009", "Tom", 55),
        ("S010", "Jerry", 75)
    ]

    for sid, name, marks in students_to_add:
        if sid not in current_student_ids:
            manager.add_student(sid, name, marks)
            current_student_ids.add(sid)
        else:
            print(f"Student with ID {sid} already present, skipping addition.")

    # Display all students
    manager.display_all()

    # Update a student's record
    print("\n--- Updating student S002 (Bob Johnson) ---")
    manager.update_student("S002", new_marks=55) # Bob now passes
    manager.update_student("S004", new_name="Tina Sharma") # Tina's name updated
    manager.display_all()

    # Remove a student's record
    print("\n--- Removing student S009 (Tom) ---")
    manager.remove_student("S009")
    manager.display_all()

    # Display only passed students
    print("\n--- Displaying only passed students ---")
    manager.show_passed()

    # Save the updated data
    print(f"\n--- Saving current student data to {filename} ---")
    manager.save_students(filename)

    # Simulate fresh start and load again to verify persistence
    print("\n--- Creating new manager instance and loading data ---")
    new_manager = StudentManager()
    new_manager.load_students(filename)

    print("\n--- Students loaded into new manager to verify persistence ---")
    new_manager.display_all()

    # Clean up the created pickle file
    if os.path.exists(filename):
        os.remove(filename)
        print(f"\nCleaned up persistence file: {filename}")

if __name__ == "__main__":
    main()
```
