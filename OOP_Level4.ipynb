```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(self.name, "-", self.marks)

    def update_marks(self, new_marks):
        self.marks = new_marks


class School:
    def __init__(self, school_name):
        self.school_name = school_name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def show_students(self):
        print("School:", self.school_name)
        for student in self.students:
            student.display()

    def average_marks(self):
        total = 0
        for student in self.students:
            total += student.marks

        if len(self.students) == 0:
            return 0

        return total / len(self.students)

    def topper(self):
        if not self.students:
            return None

        top_student = self.students[0]
        for student in self.students:
            if student.marks > top_student.marks:
                top_student = student
        return top_student

    def remove_student(self, name):
        original_len = len(self.students)
        self.students = [s for s in self.students if s.name != name]
        if len(self.students) == original_len:
            print(f"Student {name} not found in {self.school_name}.")
        else:
            print(f"Student {name} removed from {self.school_name}.")
```

```python
s1 = Student("Isha", 90)
s2 = Student("Tina", 75)

school = School("Green Valley")

school.add_student(s1)
school.add_student(s2)

school.show_students()
```

```python
print("Average:", school.average_marks())
```

```python
print("Marks before update:")
s1.display()
s2.display()

s1.update_marks(95)
s2.update_marks(80)

print("\nMarks after update:")
s1.display()
s2.display()

print("\nSchool students after update:")
school.show_students()
print("Average marks after update:", school.average_marks())
```

```python
print("Students before removal:")
school.show_students()

school.remove_student("Isha")
school.remove_student("NonExistentStudent")

print("\nStudents after removal:")
school.show_students()
print("Average marks after removal:", school.average_marks())
```
