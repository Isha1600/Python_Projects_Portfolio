```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"{self.name} - {self.marks}")

    def is_pass(self):
        return self.marks >= 40

    def grade(self):
        if self.marks >= 80:
            return "A"
        elif self.marks >= 60:
            return "B"
        elif self.marks >= 40:
            return "C"
        else:
            return "Fail"

    def update_marks(self, new_marks):
        self.marks = new_marks
```

```python
def show_grades(self):
    for student in self.students:
        print(student.name, "Grade:", student.grade())
```

```python
class Manager:
    def __init__(self, students=None):
        self.students = students if students is not None else []

    def show_grades(self):
        for student in self.students:
            print(student.name, "Grade:", student.grade())

    def find_topper(self):
        if not self.students:
            return None
        topper = self.students[0]
        for student in self.students:
            if student.marks > topper.marks:
                topper = student
        return topper
```

```python
# The show_grades method has been moved into the Manager class above.
# This cell can now be cleared or removed if it only contained that method.
```

```python
s1 = Student("Alice", 85)
s2 = Student("Bob", 30)
s3 = Student("Charlie", 65)
s4 = Student("David", 92)
s5 = Student("Eve", 40)

print("--- Initial Student Data ---")
print("Student 1:")
s1.display()
print("Student 2:")
s2.display()
print("Student 3:")
s3.display()
print("Student 4:")
s4.display()
print("Student 5:")
s5.display()
```

```python
manager = Manager([s1, s2, s3, s4, s5])
```

```python
print("\n--- Testing is_pass method ---")
for student in manager.students:
    print(f"{student.name} - Passed: {student.is_pass()}")
```

```python
print("\n--- Testing grade method ---")
for student in manager.students:
    print(f"{student.name} - Grade: {student.grade()}")
```

```python
print("\n--- Testing update_marks method ---")
print("Updating marks for Bob and Eve...")
s2.update_marks(50) # Bob now passes
s5.update_marks(75) # Eve gets a B

print("\n--- Students after updating marks ---")
for student in manager.students:
    student.display()
    print(f"  Passed: {student.is_pass()}, Grade: {student.grade()}")
```

```python
print("\n--- Testing find_topper method ---")
topper = manager.find_topper()
if topper:
    print(f"The topper is: {topper.name} with {topper.marks} marks (Grade: {topper.grade()})")
else:
    print("No students in the manager to find a topper.")
```

```python
print("\n--- Testing show_grades method (Manager) ---")
manager.show_grades()
```
