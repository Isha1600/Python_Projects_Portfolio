```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
```

```python
s1 = Student("Isha", 90)
s2 = Student("Tina", 85)

s1.display()
s2.display()
```

```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def display(self):
        print("Brand:", self.brand)
        print("Speed:", self.speed)

    def increase_speed(self, amount):
        self.speed += amount
```

```python
car1 = Car("Toyota", 120)
car2 = Car("Honda", 110)

car1.display()
car2.display()

car1.increase_speed(20)
car1.display()
```
