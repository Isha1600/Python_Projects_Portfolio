```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def update_marks(self, new_marks):
        if new_marks >= 0:
            self.marks = new_marks
        else:
            print("Marks cannot be negative.")
```

```python
s1 = Student("Isha", 90)
s1.update_marks(-50)
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks  # private

    def get_marks(self):
        return self.__marks

    def update_marks(self, new_marks):
        if new_marks >= 0:
            self.__marks = new_marks
        else:
            print("Invalid marks")
```

```python
s1 = Student("Isha", 90)
print(s1.get_marks())
```

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Invalid withdrawal")

    def show_balance(self):
        print("Balance:", self.__balance)
```

```python
acc = BankAccount("Isha", 1000)

acc.deposit(500)
acc.withdraw(200)
acc.show_balance()
acc.__balance = 1000000
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def increase_salary(self, amount):
        if amount > 0:
            self.__salary += amount
        else:
            print("Increase amount must be positive.")
```

```python
# Create an instance of the Employee class
emp = Employee("Alice", 50000)

# Get and print the initial salary
print(f"Initial salary: {emp.get_salary()}")

# Increase the salary by a positive amount
emp.increase_salary(5000)
print(f"Salary after valid increase: {emp.get_salary()}")

# Attempt to increase salary by a non-positive amount
emp.increase_salary(-2000)
print(f"Salary after invalid increase attempt: {emp.get_salary()}")
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Private method for area calculation
    def __calculate_area(self):
        return self.length * self.width

    def get_area(self):
        return self.__calculate_area()
```

```python
# Create an instance of the Rectangle class
rect = Rectangle(length=10, width=5)

# Get and print the area of the rectangle
print(f"The area of the rectangle is: {rect.get_area()}")
```
